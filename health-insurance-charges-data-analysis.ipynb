```python
fig=px.density_contour(da,x="bmi",y="charges",title = "Distribution of Insurance Charges by BMI"
)
fig.update_traces(contours_coloring="fill", contours_showlabels = True)
fig
```

```python
fig=px.density_contour(da,x="age",y="charges",title = "Distribution of Insurance Charges by Age"
)
fig.update_traces(contours_coloring="fill", contours_showlabels = True)
fig
```

```python
fig = px.scatter(da, x="age", y="charges", title = "Relationship Between Age and Insurance Charges by Gender"
,color="sex",trendline="ols",marginal_x="violin",marginal_y="violin")
fig
```

```python
fig = px.scatter(da, x="bmi", y="charges", title = "Relationship Between BMI and Insurance Charges by Smoking Status"
,color="smoker",trendline="ols",marginal_x="violin",marginal_y="violin" )
fig
```

```python
fig = px.scatter(da, x="bmi", y="charges", title = "Relationship Between BMI, Charges, and Smoking Status Across Regions"
,color="smoker",trendline="ols",marginal_x="violin",marginal_y="violin" ,facet_col="region")
fig
```

```python
fig = px.scatter(da, x="bmi", y="charges",title = "Relationship Between BMI and Insurance Charges by Gender"
,color='sex',trendline="ols",marginal_x="violin",marginal_y="violin" )
fig
```

```python
fig = px.density_heatmap(da, x="charges",title="Density Heatmap of Charges",text_auto=True)
fig
```

```python
yes_no_mapping = {'yes': 1, 'no': 0}
da['smoker'] = da['smoker'].map(yes_no_mapping)
gender_mapping={'male':1,'female':0}
da['sex']=da['sex'].map(gender_mapping)
```

# Concluding Insurance Charges Analysis

```python
df = da.select_dtypes(include=[np.int64, np.float64])
corelation=df.corr()
corelation
```

```python
fig=px.imshow(corelation,title="Correlation Matrix",text_auto=True)
fig
```

> **Hence health insurance charges are much more dependent on that whether the person is a smoker or not and age is also a contributing factor.**
